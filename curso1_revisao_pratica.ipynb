PANDAS - TRANSFORMAÇÃO E MANIPULAÇÃO DE DADOS

```python
dados = pd.read_json('/content/dados_hospedagem.json')
dados.head()
```

```python
#normalizando os dados (transformando os dados soltos em um dataframe)
dados = pd.json_normalize(dados['info_moveis'])
dados.head()
```

```python
#Transformando as colunas em uma lista
colunas = list(dados.columns)
colunas
```

```python
#desagrupando os conteudos que estavam dentro de listas nas colunas selecionadas (da terceira em diante)
dados = dados.explode(colunas[3:])
dados.head()
```

```python
#resetando os indices

dados.reset_index(drop=True, inplace=True)
dados.head()
```

```python
dados.info()
```

```python
#converter as colunas em númericas para poder trabaçhar com estatística

import numpy as np

dados['max_hospedes']=dados['max_hospedes'].astype(np.int64)
dados.info()
```

```python
col_numericas = ['quantidade_banheiros','quantidade_quartos','quantidade_camas']
dados[col_numericas] = dados[col_numericas].astype(np.int64)
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
dados.info()
```

```python
#retirar as virgulas, o cifão e os espaços em branco da coluna dinheiro
dados.info()
```

```python
dados['preco'] = dados['preco'].astype(str).str.replace('[$,]', '', regex=True).astype(np.float64)
dados.info()
```

```python
lista = ['taxa_deposito', 'taxa_limpeza']
dados[lista]=dados[lista].applymap(lambda x: x.replace(',','').replace('$','').strip())
```

```python
dados[lista] = dados[lista].astype(np.float64)
```

```python
dados.info()
```

```python
#Dividir o texto em tokens

dados['descricao_local']
```

```python
#deixar tudo em letra minuscula
dados['descricao_local'] = dados['descricao_local'].str.lower()
dados.head()
```

```python
dados['descricao_local'][3169]
```

```python
#substituição dos caracteres especiais por espaço em branco

dados['descricao_local']=dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
dados.head()
```

```python
# removendo hifens fora das palavras compostas

dados['descricao_local']=dados['descricao_local'].str.replace('(?<!\w)-(?!\w)', '', regex=True)
dados.head()
```

```python
#separar palavra por palavra atravez da virgula (tokenizar)

dados['descricao_local'] = dados['descricao_local'].str.split()
dados.head()
```

```python
#removendo '' e outros caracteres

dados['comodidades']=dados['comodidades'].str.replace('\{|}|\"', ' ', regex=True)
dados.head(3)
```

```python
#separar palavra por palavra atravez da virgula (tokenizar)

dados['comodidades'].str.split(',')
```

```python
dados['comodidades']=dados['comodidades'].str.split(',')
```

```python
dados.head(3)
```
